```python
import random

random.seed()
```

```python
random.random()
```

```python
n = []
for i in range(0,10):
    n.append(random.random())
n
```

```python
n = [random.random() for _ in range(0,10)]
n
```

```python
import matplotlib.pyplot as plt

plt.scatter(range(0,10), n)
plt.show()
```

```python
def numbers(n,m=1.0):
    return [random.random()*m for _ in range(0,n)]

numbers(3)
```

```python
plt.scatter(numbers(10000,2), numbers(10000,2))
plt.show()
```

```python
import math

def distance(x0,y0,x1,y1):
    return math.sqrt((x0-x1)*(x0-x1) + (y0-y1)*(y0-y1))

distance(1,1,2,2)
```

```python
p = list(zip([1,2,3],[4,5,6]))
p
```

```python
p2 = [(x,y) for (x,y) in p if x + y > 8]
p2
```

```python
points = list(zip(numbers(1000000,2),numbers(1000000,2)))
len(points),points[0:5]
```

```python
plt.scatter([x for (x,y) in points], [y for (x,y) in points])
plt.show()
```

```python
points_in_circle = [(x,y) for (x,y) in points if distance(x,y,1,1) <= 1]
```

```python
plt.scatter([x for (x,y) in points_in_circle], [y for (x,y) in points_in_circle])
plt.show()
```
